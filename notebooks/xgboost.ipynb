```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df= pd.read_csv("train.csv")

df.head()
```

```python
df = pd.get_dummies(
    df,
    columns=["Barrio", "Categoria_Seguridad"],
    drop_first=False,
    dtype=int
)

df.to_csv("train.csv")
```

```python
from sklearn.model_selection import train_test_split, ParameterSampler
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, mean_absolute_percentage_error
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.inspection import permutation_importance
from xgboost import XGBRegressor
import time

RANDOM_STATE = 55
TARGET = "Precio"
OUTPUT_PATH = "predicciones_xgboost.csv"

df_model = df.copy()

X = df_model.drop(columns=[TARGET, "Precio_m2", "Percentil_Precio_Barrio"], errors="ignore")
y = df_model[TARGET]

cat_cols = X.select_dtypes(include=["object", "category"]).columns.tolist()
num_cols = X.select_dtypes(include=np.number).columns.tolist()

try:
    ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)
except TypeError:
    ohe = OneHotEncoder(handle_unknown="ignore", sparse=False)

preprocessor = ColumnTransformer(
    transformers=[
        ("num", SimpleImputer(strategy="median"), num_cols),
        ("cat", Pipeline([
            ("imputer", SimpleImputer(strategy="most_frequent")),
            ("encoder", ohe)
        ]), cat_cols)
    ]
)

X_fit, X_valid, y_fit, y_valid = train_test_split(
    X, y, test_size=0.20, random_state=RANDOM_STATE
)

base_xgb_params = dict(
    objective="reg:squarederror",
    random_state=RANDOM_STATE,
    n_jobs=-1,
    tree_method="hist"
)

param_distributions = {
    "n_estimators": [500, 800, 1000, 1200],
    "max_depth": [2, 3, 4, 5, 6],
    "learning_rate": [0.01, 0.03, 0.05, 0.08],
    "subsample": [0.7, 0.8, 0.9, 1.0],
    "colsample_bytree": [0.7, 0.8, 0.9, 1.0],
    "min_child_weight": [1, 3, 5, 10, 20],
    "reg_alpha": [0, 0.01, 0.1, 0.5, 1],
    "reg_lambda": [1, 3, 5, 10, 20]
}

param_list = list(ParameterSampler(
    param_distributions=param_distributions,
    n_iter=50,
    random_state=RANDOM_STATE
))

best_rmse = np.inf
best_params = None
results = []
start_time = time.time()

for i, params in enumerate(param_list, start=1):
    iter_start = time.time()

    model = Pipeline([
        ("preprocessor", preprocessor),
        ("model", XGBRegressor(**base_xgb_params, **params))
    ])

    model.fit(X_fit, y_fit)

    pred_fit = model.predict(X_fit)
    pred_valid = model.predict(X_valid)

    r2_fit = r2_score(y_fit, pred_fit)
    r2_valid = r2_score(y_valid, pred_valid)
    r2_gap = r2_fit - r2_valid

    rmse_fit = np.sqrt(mean_squared_error(y_fit, pred_fit))
    rmse_valid = np.sqrt(mean_squared_error(y_valid, pred_valid))

    mae_fit = mean_absolute_error(y_fit, pred_fit)
    mae_valid = mean_absolute_error(y_valid, pred_valid)

    mape_fit = mean_absolute_percentage_error(y_fit, pred_fit) * 100
    mape_valid = mean_absolute_percentage_error(y_valid, pred_valid) * 100

    if rmse_valid < best_rmse:
        best_rmse = rmse_valid
        best_params = params.copy()
        marker = " <-- best"
    else:
        marker = ""

    results.append({
        **params,
        "r2_fit": r2_fit,
        "r2_valid": r2_valid,
        "r2_fit_valid_gap": r2_gap,
        "rmse_fit": rmse_fit,
        "rmse_valid": rmse_valid,
        "mae_fit": mae_fit,
        "mae_valid": mae_valid,
        "mape_fit": mape_fit,
        "mape_valid": mape_valid,
        "iter_time_sec": time.time() - iter_start,
        "elapsed_sec": time.time() - start_time
    })

    print(
        f"[{i:03d}/{len(param_list)}] "
        f"rmse_valid={rmse_valid:.2f} "
        f"mae_valid={mae_valid:.2f} "
        f"mape_valid={mape_valid:.2f}% "
        f"r2_valid={r2_valid:.5f} "
        f"r2_fit={r2_fit:.5f} "
        f"gap={r2_gap:.5f} "
        f"best_rmse={best_rmse:.2f} "
        f"params={params}{marker}"
    )

results_xgboost = pd.DataFrame(results).sort_values("rmse_valid", ascending=True)

display(results_xgboost.head(20))

print("Best params:")
print(best_params)
print("Best validation RMSE:")
print(best_rmse)
print("Best validation R2:")
print(results_xgboost.iloc[0]["r2_valid"])
print("Best validation MAE:")
print(results_xgboost.iloc[0]["mae_valid"])
print("Best validation MAPE:")
print(results_xgboost.iloc[0]["mape_valid"])

best_model_valid = Pipeline([
    ("preprocessor", preprocessor),
    ("model", XGBRegressor(**base_xgb_params, **best_params))
])

best_model_valid.fit(X_fit, y_fit)

feature_names = best_model_valid.named_steps["preprocessor"].get_feature_names_out()
feature_names = [x.replace("num__", "").replace("cat__", "") for x in feature_names]

feature_importance = pd.DataFrame({
    "Feature": feature_names,
    "Importance": best_model_valid.named_steps["model"].feature_importances_
}).sort_values("Importance", ascending=False)

display(feature_importance.head(30))

plt.figure(figsize=(10, 8))
sns.barplot(data=feature_importance.head(30), x="Importance", y="Feature")
plt.title("Feature Importance - XGBoost")
plt.show()

perm = permutation_importance(
    best_model_valid,
    X_valid,
    y_valid,
    n_repeats=10,
    random_state=RANDOM_STATE,
    scoring="neg_root_mean_squared_error",
    n_jobs=-1
)

permutation_importance_df = pd.DataFrame({
    "Feature": X_valid.columns,
    "Importance_Mean": perm.importances_mean,
    "Importance_Std": perm.importances_std
}).sort_values("Importance_Mean", ascending=False)

display(permutation_importance_df.head(30))

plt.figure(figsize=(10, 8))
sns.barplot(data=permutation_importance_df.head(30), x="Importance_Mean", y="Feature")
plt.title("Permutation Importance - XGBoost")
plt.show()

best_model = Pipeline([
    ("preprocessor", preprocessor),
    ("model", XGBRegressor(**base_xgb_params, **best_params))
])

best_model.fit(X, y)

df_pred = df.copy()
df_pred["Prediccion_Precio"] = best_model.predict(X)

df_pred.to_csv(OUTPUT_PATH, index=False)

print("Guardado:", OUTPUT_PATH)
```

### Conclusión

El modelo XGBoost le da mayor importancia al barrio, especialmente Puerto Madero, que aparece como la variable más influyente. Después, las variables más relevantes son Expensas**, Cochera, Superficie cubierta, Baños** y Superficie total, lo que indica que el precio se explica principalmente por ubicación, tamaño y características estructurales del departamento.

Las variables de seguridad, cercanía al subte y apto crédito tienen menor peso relativo en comparación con las variables principales.


La permutation importance muestra que el modelo depende principalmente de variables estructurales del departamento: superficie cubierta, superficie total, antigüedad y expensas. Esto indica que, al alterar esas variables, el error del modelo aumenta mucho, por lo que son claves para predecir el precio.

A diferencia de la importancia interna de XGBoost, acá Puerto Madero sigue siendo relevante, pero no aparece como la variable dominante. Las variables de barrio, amenities y cercanía al subte tienen menor impacto relativo.
